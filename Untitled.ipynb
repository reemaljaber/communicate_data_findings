```python
#removing some unnessacry details about Occupations to make them easier for visualizations and statistical analysis

Occup_abbrev = {'Student - Technical School':'Student',
               'Student - Community College':'Student',
               'Student - College Freshman':'Student',
               'Student - College Sophomore':'Student',
               'Student - College Junior':'Student',
               'Student - College Senior':'Student',
               'Student - College Graduate Student':'Student',
               'Nurse (RN)':'Nurse',
               'Nurse (LPN)':'Nurse',
               "Nurse's Aide":'Nurse',
               'Tradesman - Mechanic':'Tradesman',
               'Tradesman - Electrician':'Tradesman',
               'Tradesman - Carpenter':'Tradesman',
               'Tradesman - Plumber':'Tradesman',
               'Engineer - Mechanical':'Engineer',
               'Engineer - Chemical':'Engineer',
               'Engineer - Electrical':'Engineer',
               'Sales - Commission':'Sales',
               'Sales - Retail':'Sales',
               'Police Officer/Correction Officer':'Officer',
               'Military Officer':'Officer',
               'Food Service Management':'Food Service',
               'Clergy':'Religious',
               "Teacher's Aide":'Teacher'}
def abbreviate_occup(df2):
    
    if df2['Occupation'] in Occup_abbrev :
        abbrev = Occup_abbrev[df2['Occupation']]
        return abbrev
    else :
        return df2['Occupation']


df2['Occupation'] = df2.apply(abbreviate_occup,axis=1)
```
