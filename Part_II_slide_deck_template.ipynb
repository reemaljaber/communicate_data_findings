## Distribution of Debt To Income Ratio

these two plots are showing us the average ratio of debt to income that usully people have.
on the debt to income ratio plot most of the data sets of the far left of the axe between zero and two, the distribution of  the debt to income ratio takes on a unimodal shap but slightly skeawed, so i had to investigate further on an even smaller bin size.
as we can see that the majorty have their debt ratio between 0.1 and 0.3 of their income. 

```python
sb.histplot(df['DebtToIncomeRatio'])
plt.xlabel('DebtToIncomeRatio')
plt.ylabel('Count ')
plt.title('Distribution of DebtToIncomeRatio')
plt.show()
```

```python
sb.histplot(df['DebtToIncomeRatio'])
plt.xlim([0,0.7])
plt.xlabel('Debt To Income Ratio')
plt.ylabel('count')
plt.title('Distribution of Debt To Income Ratio')
plt.show()
```

## log scatter plot of Borrower APR vs. Borrower Rate

BorrowerAPR had a high amount of correlation with the BorrowerRate, even before transforming the features. An approximately linear relationship was observed when BorrowerAPR was Plotted on a logarithmic scale.

```python
plt.figure(figsize = [8, 6])
plt.scatter(data = df, x = 'BorrowerRate', y = 'BorrowerAPR',alpha = 5/10)
plt.xlabel('Borrower Rate')
plt.yscale('log')
plt.yticks([0.006, 0.01, 0.02, 0.05, 0.1, 0.2,0.4,0.8], [0.006, 0.01, 0.02, 0.05, 0.1, 0.2,0.4,0.8])
plt.ylabel('Borrower APR')
plt.title('log scatter plot of Borrower APR vs. Borrower Rate')
plt.show()
```
